### Shakespeare Word Frequency

```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.gutenberg.org/files/1513/1513-h/1513-h.htm')
html_string = response.text
document = BeautifulSoup(html_string,"html.parser")
```

```python
print(document.text)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
# tokenizing the words
print(word_tokenize(document.text))
```

```python
words = []
for w in word_tokenize(document.text):
        words.append(w)
```

```python
print(words)
```

```python
print(stopwords.words('english'))
```

```python
print(list(punctuation))
```

```python
myStopWords = list(punctuation) + stopwords.words('english')
```

```python
print(myStopWords)
```

```python
wordsNoStop = []
for i in words:
    if i not in myStopWords:
        wordsNoStop.append(i)
print(wordsNoStop)
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(wordsNoStop)
```

```python
# Top 20 most frequent words of the play
for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

##### The top 20 most frequent words of the play do not seem to give an accurate sense of the play beside the words 'Romeo','love','thee','Juliet'. If it was possible to add stopwords and punctuation to the NLTK library then we might get a better list of words that give us a better sense of the play.

```python
# Creating a python list that contains all ten strings
listOfMovies = [one,two,three,four,five,six,seven,eight,nine,ten]
```

```python
listOfMovies
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
extrastop = ['``',"''","'re","'s","'re",'``',"''","'ll","--","\'\'","...",
             "n\'t",'one','would','use','subject','from',"\'m","\'ve","good","great","enjoy","film"]
myStopWords = list(punctuation) + stopwords.words('english') + extrastop
```

```python
# tokenizing words from string list while removing all the stop words
listOfMoviesWords = []
for i in listOfMovies:
    listOfMoviesWords.append([w for w in word_tokenize(i.lower()) if w not in myStopWords])
```

```python
listOfMoviesWords
```

```python
from nltk.stem.porter import PorterStemmer
p_stemmer = PorterStemmer()
```

```python
# stemming the tokenized words in the movies list
listOfStemmedWords = []
for i in listOfMoviesWords:
    listOfStemmedWords.append([p_stemmer.stem(w) for w in i])
```

```python
!pip install gensim
```

```python
from gensim import corpora, models
import gensim
```

```python
dictionary = corpora.Dictionary(listOfStemmedWords)
corpus = [dictionary.doc2bow(text) for text in listOfStemmedWords]
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=2, 
                                           id2word = dictionary, 
                                           passes=20)
```

```python
ldamodel.print_topics(num_topics=2, num_words=20)
```

#### It seems like the model did a good job at representing the comedy review topic in both splits of topics by identifying words such as 'joke','comedi', and 'fun'. However it didnt really pick up the topic of drama movie reviews, the closest words to the topic it picked up were 'challeng','power' and 'shadow'. Overall, from these two models I am able to see that a comedy topic was analyzed but I am not able to see that there was a drama topic involved.
