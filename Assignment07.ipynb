```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.gutenberg.org/files/1513/1513-h/1513-h.htm')
html_string = response.text
document = BeautifulSoup(html_string,"html.parser")
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
# tokenizing the words
print(word_tokenize(document.text))
```

```python
words = []
for w in word_tokenize(document.text):
        words.append(w)
```

```python
print(words)
```

```python
extrastop = ['``',"''","'re","'s","'re",'``',"''","'ll","--","\'\'","...",
             "n\'t",'one','would','use','subject','from',"\'m","\'ve","'",","]
myStopWords = list(punctuation) + stopwords.words('english') + extrastop
```

```python
# Removing Stopwords
wordsNoStop = []
for i in words:
    if i not in myStopWords:
        wordsNoStop.append(i)
print(wordsNoStop)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Stemming the words
Stemmedwords = []
for word in wordsNoStop:
    Stemmedwords.append(ps.stem(word))
```

```python
Stemmedwords
```

```python
from nltk.probability import FreqDist
```

```python
# Getting word frequency from stemmed word list
freq = FreqDist(Stemmedwords)
```

```python
freq
```

```python
import pandas as pd
```

```python
# Turning FreqDist list into pandas dataframe
df = pd.DataFrame.from_dict(freq, orient = 'index')
df.columns = ['Frequency']
df.index.name = 'Word'
print(df)
```

```python
# finding the mean of word frequencies
df.mean()
```

```python
# finding the median of word frequencies
df.median()
```

```python
# finding the mode of word frequencies
df.mode()
```

```python
from scipy import stats
```

```python
# Finding the 10% trimmed mean of word frequencies 
# I used the SciPy library just because its commands seemed easier to use for finding the trimmed mean.
stats.trim_mean(df.Frequency, 0.1)
```

```python
df.plot(kind='hist',bins=30)
print('Mean: ',df.mean())
print('Median: ',df.median())
print('Mode: ',df.mode())
print('Trimmed Mean: ',stats.trim_mean(df.Frequency, 0.1))
```

Its very hard to distinguish the mean median and mode from this histogram and I think that it is because there are many words that have a frequency of 1 in the list.

```python
# Finding the standard deviation of word frequencies
df.std()
```

```python
# Calculating the interquartile range 
df.quantile(.75) - df.quantile(.25)
```

There seems to be a big gap between the standard deviation and the interquartile range of word frequencies. Which makes sense because in the histogram there is a large range that can be seen, this aligns with the large standard deviation of 24. The interquartile range is 3 which lines up with the mean of 5 for the word frequencies however, given the skewed graph it is hard to identify on the chart.

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = df[['s5']]
y = df['target']
```

```python
reg = LinearRegression().fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
ytrain = reg.intercept_ + reg.coef_ * X
```

```python
# Plotting the linear fit on top of the scatter plot
plt.plot(X,y,'ro',X,ytrain,'b-');
```

```python
mean_squared_error(y, ytrain)
```

```python
r2_score(y, ytrain)
```

```python
# Doing linear regression with another variable 
X = df[['s3']]
y = df['target']
```

```python
reg = LinearRegression().fit(X, y)
```

```python
ytrain = reg.intercept_ + reg.coef_ * X
```

```python
mean_squared_error(y, ytrain)
```

```python
r2_score(y, ytrain)
```

The R squared value has decreased and the mean squared error has increased when doing a linear regression between high-density lipoproteins and the target values compared to doing it between triglycerides levels and the target values, which makes sense because high-density lipoprotein levels do not have a big impact on the progression of diabetes, compared to triglyceride levels.
